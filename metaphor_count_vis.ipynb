```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.style.use("fivethirtyeight")
```

## Bhagavad Gita count plots

```python
mdf = pd.read_csv('/content/drive/MyDrive/xlmr_mg_metaphor.csv')
edf = pd.read_csv('/content/drive/MyDrive/xlmr_ee_metaphor.csv')
sdf = pd.read_csv('/content/drive/MyDrive/xlmr_sps_metaphor.csv')
```

```python
edf[edf['chapter']==1]
```

```python
mdf['text_count'] = mdf['text'].apply(lambda x: len(x.split()))
edf['text_count'] = edf['text'].apply(lambda x: len(x.split()))
sdf['text_count'] = sdf['text'].apply(lambda x: len(x.split()))
```

```python
mgk = mdf.groupby('chapter').sum().reset_index()
egk = edf.groupby('chapter').sum().reset_index()
sgk = sdf.groupby('chapter').sum().reset_index()
mgk
```

```python
import os
if os.path.exists('/content/images')!=True:
  os.mkdir('/content/images')
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(16,5))
p6=sns.lineplot(x = mgk['chapter'], y = mgk['count_met_ref'], label='Mahatama Gandhi')
p6.set_xlabel("Mahatama Gandhi Translations")
p6=sns.lineplot(x = egk['chapter'], y = egk['count_met_ref'], label='Eknath Easwaran')
p6.set_xlabel("Eknath Easwaran Translations")
p6=sns.lineplot(x = sgk['chapter'], y = sgk['count_met_ref'], label='Shri Swami Purohit')
p6.set_xlabel("Shri Swami Purohit Translations")

plt.ylabel('Metaphors Count', labelpad=40, fontsize=30)
plt.xlabel('Chapter', labelpad=30, fontsize=30)

# plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))    # format decimals on the y-axis
plt.yticks(fontsize= 25)
plt.xticks(np.arange(1, 19, step=1), fontsize = 25)

plt.savefig('/content/images/bg_all_count.png', bbox_inches = 'tight')
plt.show()
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(16,5))
p6=sns.lineplot(x = mgk['chapter'], y = mgk['count_verb_ref'], label='Mahatama Gandhi')
p6.set_xlabel("Mahatama Gandhi Translations")
p6=sns.lineplot(x = egk['chapter'], y = egk['count_verb_ref'], label='Eknath Easwaran')
p6.set_xlabel("Eknath Easwaran Translations")
p6=sns.lineplot(x = sgk['chapter'], y = sgk['count_verb_ref'], label='Shri Swami Purohit')
p6.set_xlabel("Shri Swami Purohit Translations")

plt.ylabel('Verb Metaphors Count', labelpad=40, fontsize=30)
plt.xlabel('Chapter', labelpad=30, fontsize=30)

# plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))    # format decimals on the y-axis
plt.yticks(fontsize= 25)
plt.xticks(np.arange(1, 19, step=1), fontsize = 25)

plt.savefig('/content/images/bg_vb_count.png', bbox_inches = 'tight')
plt.show()
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(16,5))
p6=sns.lineplot(x = mgk['chapter'], y = mgk['text_count'], label='Mahatama Gandhi')
p6.set_xlabel("Mahatama Gandhi Translations")
p6=sns.lineplot(x = egk['chapter'], y = egk['text_count'], label='Eknath Easwaran')
p6.set_xlabel("Eknath Easwaran Translations")
p6=sns.lineplot(x = sgk['chapter'], y = sgk['text_count'], label='Shri Swami Purohit')
p6.set_xlabel("Shri Swami Purohit Translations")

plt.ylabel('Word Count', labelpad=40, fontsize=30)
plt.xlabel('Chapter', labelpad=30, fontsize=30)

# plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))    # format decimals on the y-axis
plt.yticks(fontsize= 25)
plt.xticks(np.arange(1, 19, step=1), fontsize = 25)

plt.savefig('/content/images/bg_text_count.png', bbox_inches = 'tight')
plt.show()
```

## Sermon on the mount

```python
kdf = pd.read_csv('/content/drive/MyDrive/xlmr_kj.csv')
ndf = pd.read_csv('/content/drive/MyDrive/xlmr_ni.csv')
rdf = pd.read_csv('/content/drive/MyDrive/xlmr_rs.csv')
kdf['text_count'] = kdf['verse'].apply(lambda x: len(x.split()))
ndf['text_count'] = ndf['verse'].apply(lambda x: len(x.split()))
rdf['text_count'] = rdf['verse'].apply(lambda x: len(x.split()))
kgk = kdf.groupby('chapter').sum().reset_index()
ngk = ndf.groupby('chapter').sum().reset_index()
rgk = rdf.groupby('chapter').sum().reset_index()
kgk
```

```python
plt.rcParams.update({'font.size': 10})
plt.figure(figsize=(8,4))
p6=sns.lineplot(x = kgk['chapter'], y = kgk['count_met_ref'], label='King James Version')
p6.set_xlabel("King James Translations")
p6=sns.lineplot(x = ngk['chapter'], y = ngk['count_met_ref'], label='New International Version')
p6.set_xlabel("New International Translations")
p6=sns.lineplot(x = rgk['chapter'], y = rgk['count_met_ref'], label='New Revised Standard Version')
p6.set_xlabel("New Revised Standard Translations")

plt.ylabel('Metaphors Count', labelpad=10, fontsize=20)
plt.xlabel('Chapter', labelpad=5, fontsize=20)

# plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))    # format decimals on the y-axis
plt.yticks(fontsize= 18)
plt.xticks(np.arange(5, 8, step=1), fontsize = 18)

plt.savefig('/content/images/sm_all_count.png', bbox_inches = 'tight')
plt.show()
```

```python
plt.rcParams.update({'font.size': 10})
plt.figure(figsize=(8,4))
p6=sns.lineplot(x = kgk['chapter'], y = kgk['count_verb_ref'], label='King James Version')
p6.set_xlabel("King James Translations")
p6=sns.lineplot(x = ngk['chapter'], y = ngk['count_verb_ref'], label='New International Version')
p6.set_xlabel("New International Translations")
p6=sns.lineplot(x = rgk['chapter'], y = rgk['count_verb_ref'], label='New Revised Standard Version')
p6.set_xlabel("New Revised Standard Translations")

plt.ylabel('Verb Metaphors Count', labelpad=10, fontsize=20)
plt.xlabel('Chapter', labelpad=5, fontsize=20)

# plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))    # format decimals on the y-axis
plt.yticks(fontsize= 18)
plt.xticks(np.arange(5, 8, step=1), fontsize = 18)

plt.savefig('/content/images/sm_vb_count.png', bbox_inches = 'tight')
plt.show()
```
